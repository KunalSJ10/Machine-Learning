```python
import pandas as pd
df = pd.read_csv("titanic.csv")
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df.head()
```

```python
inputs = df.drop('Survived',axis='columns')
target = df.Survived
```

```python
dummies = pd.get_dummies(inputs.Sex)
dummies = dummies.astype(int)
dummies.head(3)
```

```python
inputs = pd.concat([inputs,dummies],axis='columns')
inputs.head(3)
```

```python
inputs.drop(['Sex','male'],axis='columns',inplace=True)
inputs.head(3)
```

```python
'''inputs.isnull().values.any()
#check in whole dataset
inputs['Age'].isnull().any()
#check in that specific columns
inputs[inputs.isnull().any(axis=1)]
#list of all dataset with null
inputs[inputs['Age'].isnull()]
#for specific row'''
inputs.columns[inputs.isna().any()]
#give the columns with null value
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.3)
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict(X_test[0:10])
```

```python
model.predict_proba(X_test[:10])
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(GaussianNB(),X_train, y_train, cv=5)
```
